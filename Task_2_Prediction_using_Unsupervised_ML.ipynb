### **Import Libraries**

```python
# Imported all the required libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn import datasets
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')
```

### **Reading the dataset**

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

### **Exploring the data and checking the null values**

```python
# Checking the null values
df.isnull()
```

### **Displaying all the columns and their data types**

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

### **Checking the missing values**

```python
df.isnull().sum().sort_values(ascending = False)
```

### **Checking duplicate values**

```python
df.duplicated().sum()
```

### **Drop the duplicate values**

```python
df.drop_duplicates(inplace = True)
df
```

```python
df.shape
```

### **Checking for the outliers**

```python
for i in df.columns:
    if df[i].dtype == 'float64':
        plt.figure(figsize = (5, 3))
        sns.boxplot(df[i])
        plt.show()
```

### **Targeting the Outliers**

```python
Q1, Q3 = np.percentile(df['sepal width (cm)'], [25, 75])

IQR = Q3 - Q1
lower = Q1 - (1.5 * IQR)
upper = Q3 + (1.5 * IQR)

df = df[(df['sepal width (cm)'] >= lower) & (df['sepal width (cm)'] <= upper)]
df.shape
```

```python
# Boxplot for sepal width (cm) after outlier treatment
plt.figure(figsize = (5, 3))
sns.boxplot(y = df['sepal width (cm)'])
plt.show()
```

### **Correlation Matrix**

```python
df.corr()
```

```python
plt.figure(figsize = (10, 7))
sns.heatmap(abs(df.corr()), cmap = 'crest', annot = True)
plt.show()
```

### **Find the optimum numbers of K - clusters**

```python
# creating X to place into clusters
x = df.iloc[:, [0, 1, 2, 3]].values

# defining and plotting the Elbow method
# function to calculate WCSS
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### **Applying kmeans to the dataset / Creating the kmeans classifier**

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

### **Visualising the clusters - On the first two columns**

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'purple', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```

### **Plotting the centroids of the clusters**

```python
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```

```python
df['cluster'] = kmeans.labels_
df
```
